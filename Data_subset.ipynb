```python
#Boolean for missing data
Missing=ship.isna()
Missing
```

```python
#Count total no of missing values in each column
Missing.sum()
```

```python
#percentage of missing data
Miss_data= (Missing.sum()/len(ship))*100
Miss_data
```

```python
#percentage round off
print(round(Miss_data,2))
```

```python
#data for which age is missing 
Age_missing=ship[ship["Age"].isna()]
Age_missing
```

```python
#Data for which age is not null
age=ship[ship["Age"].notna()]
age
```

```python
age.shape
```

```python
#data for which age and cabin are not null
Age_cabin=ship[(ship['Age'].notna()) & (ship['Cabin'].notna())]
Age_cabin
```

# Select subset of Given data

```python
#We want only name, age ,sex and cabin from the given data
age=ship[["Name",'Age','Sex','Cabin']]
age
```

```python
#Data for which age grater than or equal to 45
age_45=ship[ship["Age"]>=45]
age_45
```

```python
age_45.shape
```

```python
#want data for passenger whose Pclass was 2 or 3
class_2and3 = ship[ship["Pclass"].isin([2, 3])]
class_2and3
```

```python
#Data subset for Multiple conditions at a time
Multi_condn=ship[(ship["Age"].notna()) & (ship["Pclass"].isin([2])) & (ship["Sex"]=='male')]
Multi_condn
```

```python
Multi_condn.shape
```

```python
Multi_condn.info()
```

```python
# subset of both rows and columns at a time
adult_name=ship.loc[ship["Age"]>25,["Name","Age","Sex"]]
adult_name
```
