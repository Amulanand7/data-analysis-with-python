```python
#Making first column as 0th index
airq=pd.read_csv("data/air_quality_no2.csv",index_col=0,parse_dates=True)
```

```python
airq.plot()
mat.title("Air Quality").set_color("purple")
mat.xlabel('Dates').set_color('Red') 
mat.ylabel('Values').set_color('blue')
```

```python
airq.plot.box()
```

```python
airq.plot.hexbin('station_paris','station_london',gridsize=(10,10))
```

```python
airq.plot.area(figsize=(12,6))
```

```python
airq.plot.area(figsize=(10,4))
```

```python
airq.plot.area(figsize=(12,4),subplots=True)
```

```python
#setting up parameter,title and label with name and colors
airq["station_antwerp"].plot()
mat.tick_params(axis='x', colors='black')
mat.tick_params(axis='y', colors='purple')
mat.title("Antwerp").set_color('blue')
mat.ylabel('Values').set_color('Orange')
mat.xlabel('Dates').set_color('red')
```

```python
airq["station_paris"].plot()
mat.title("Pairs")
mat.tick_params()
```

```python
airq["station_london"].plot()
mat.title("London").set_color("Red")
```

```python
#comparing air quality of two cities
airq.plot.scatter(x="station_london",y="station_paris",color=['red'],alpha=0.3)
mat.title('Paris Vs London').set_color("red")
```
